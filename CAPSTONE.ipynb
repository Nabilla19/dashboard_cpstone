###Melakukan Exploratory Data Analysis (EDA) untuk mendapatkan insight dari data.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

**1. Distribusi Risk Level**

````python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='risk_level'
)

plt.title('Distribusi Tingkat Risiko Mental')
plt.xlabel('Risk Level')
plt.ylabel('Jumlah Mahasiswa')

plt.show()
````

**Insight**

Sebagian besar mahasiswa berada pada kategori risiko rendah, namun terdapat juga mahasiswa dengan risiko sedang dan tinggi yang perlu diperhatikan.

**2. Exam Pressure vs Stress Level**

````python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='exam_pressure',
    y='stress_level'
)

plt.title('Hubungan Tekanan Akademik dan Tingkat Stres')
plt.xlabel('Exam Pressure')
plt.ylabel('Stress Level')

plt.show()
````

**Insight**

Semakin tinggi tekanan akademik, tingkat stres mahasiswa cenderung meningkat.

**3. Sleep Hours vs Stress Level**

````python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='sleep_hours',
    y='stress_level'
)

plt.title('Hubungan Jam Tidur dan Tingkat Stres')
plt.xlabel('Sleep Hours')
plt.ylabel('Stress Level')

plt.show()
````

**Insight**

Mahasiswa dengan jam tidur lebih rendah cenderung memiliki tingkat stres yang lebih tinggi

**4. Social Support vs Depression Score**

````python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='social_support',
    y='depression_score'
)

plt.title('Pengaruh Dukungan Sosial terhadap Depresi')
plt.xlabel('Social Support')
plt.ylabel('Depression Score')

plt.show()
````

**Insight**

Dukungan sosial yang tinggi cenderung menurunkan tingkat depresi mahasiswa

1. Lihat Skala Warna (Legenda di Kanan)
Warna memberi tahu kamu arah dan kekuatan hubungan:

Merah (Positif): Jika satu variabel naik, variabel satunya ikut naik. Makin merah (mendekati +1.00), makin kuat hubungannya.

Biru (Negatif): Jika satu variabel naik, variabel satunya malah turun. Makin biru (mendekati -1.00), makin kuat hubungan terbaliknya.

Abu-abu/Putih (Nol): Tidak ada hubungan sama sekali antar kedua variabel tersebut.

2. Mencari Hubungan Paling Menarik
Berdasarkan gambar yang kamu bagikan, ada beberapa poin yang "teriak" untuk diperhatikan:

Korelasi Sangat Kuat (Warna Merah Pekat):

Screen Time & Internet Usage (0.89): Ini masuk akal. Makin lama orang pakai internet, makin tinggi waktu layarnya.

Stress Level & Burnout Score (0.75): Ada hubungan kuat antara tingkat stres yang tinggi dengan risiko kelelahan mental (burnout).

Study Hours & Exam Pressure (0.77): Makin banyak jam belajar, biasanya tekanan ujian yang dirasakan juga makin besar.

Korelasi Terbalik/Negatif (Warna Biru Pekat):

Stress Level & Mental Health Index (-0.95): Ini korelasi yang sangat ekstrem. Artinya, makin tinggi tingkat stres seseorang, kesehatan mentalnya turun drastis. Tanda minus (-) menunjukkan hubungan yang berlawanan arah.

Burnout Score & Mental Health Index (-0.80): Sama seperti stres, kalau skor burnout tinggi, indeks kesehatan mental biasanya rendah.
