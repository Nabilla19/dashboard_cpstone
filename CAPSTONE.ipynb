**7. Correlation Heatmap**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Menghitung korelasi antar fitur numerik
corr = df.corr(numeric_only=True)

# Membuat mask untuk menyembunyikan segitiga atas
mask = np.triu(np.ones_like(corr, dtype=bool))

# Membuat heatmap
plt.figure(figsize=(16, 10))

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    annot_kws={"size": 8},
    cbar_kws={"shrink": 0.8}
)

# Judul dan tampilan label
plt.title('Correlation Heatmap', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
````

. Lihat Skala Warna (Legenda di Kanan)
Warna memberi tahu kamu arah dan kekuatan hubungan:

Merah (Positif): Jika satu variabel naik, variabel satunya ikut naik. Makin merah (mendekati +1.00), makin kuat hubungannya.

Biru (Negatif): Jika satu variabel naik, variabel satunya malah turun. Makin biru (mendekati -1.00), makin kuat hubungan terbaliknya.

Abu-abu/Putih (Nol): Tidak ada hubungan sama sekali antar kedua variabel tersebut.

# FEATURE ENGINEERING

````python
# Sleep Category
df['sleep_category'] = df['sleep_hours'].apply(
    lambda x: 'Kurang' if x < 6 else 'Cukup'
)

# Screen Time Category
df['screen_time_category'] = df['screen_time'].apply(
    lambda x: 'Tinggi' if x > 8 else 'Normal'
)

# Stress Category
def stress_category(x):
    if x <= 3:
        return 'Low'
    elif x <= 7:
        return 'Medium'
    else:
        return 'High'

df['stress_category'] = df['stress_level'].apply(stress_category)

# Mental Risk Score
df['mental_risk_score'] = (
    df['stress_level'] +
    df['anxiety_score'] +
    df['depression_score']
) / 3

# Social Support Category
df['support_category'] = df['social_support'].apply(
    lambda x: 'Low Support' if x < 5 else 'High Support'
)

# Melihat hasil
df.head()
````

````python
df.to_csv('mental_health_featured.csv', index=False)
````

**Feature Engineering**

Pada tahap feature engineering dilakukan pembuatan fitur baru dari data yang sudah dibersihkan untuk menghasilkan informasi yang lebih mudah dipahami dan lebih informatif bagi model.

### 1. Sleep Category
Fitur `sleep_category` dibuat berdasarkan jumlah jam tidur mahasiswa (`sleep_hours`).

Kriteria:
- < 6 jam → Kurang
- ≥ 6 jam → Cukup

Tujuan:
Mengelompokkan pola tidur mahasiswa agar lebih mudah dianalisis terhadap kondisi kesehatan mental.

---

### 2. Screen Time Category
Fitur `screen_time_category` dibuat berdasarkan durasi penggunaan layar (`screen_time`).

Kriteria:
- > 8 jam → Tinggi
- ≤ 8 jam → Normal

Tujuan:
Mengidentifikasi hubungan penggunaan perangkat dengan kondisi mental mahasiswa.

---

### 3. Stress Category
Fitur `stress_category` dibuat berdasarkan tingkat stres (`stress_level`).

Kriteria:
- 0–3 → Low
- 4–7 → Medium
- 8–10 → High

Tujuan:
Mengelompokkan tingkat stres agar lebih mudah dibaca dan dianalisis.

---

### 4. Mental Risk Score
Fitur `mental_risk_score` dibuat dari rata-rata:

- stress_level
- anxiety_score
- depression_score

Rumus:

Mental Risk Score = (Stress + Anxiety + Depression) / 3

Tujuan:
Membuat indikator gabungan yang merepresentasikan kondisi kesehatan mental mahasiswa.

---

### 5. Support Category
Fitur `support_category` dibuat berdasarkan tingkat dukungan sosial (`social_support`).

Kriteria:
- < 5 → Low Support
- ≥ 5 → High Support

Tujuan:
Melihat pengaruh dukungan sosial terhadap kesehatan mental mahasiswa.

---

Kesimpulan:

Feature engineering dilakukan untuk menghasilkan fitur yang lebih informatif sehingga dapat membantu model dalam memahami pola kesehatan mental mahasiswa dengan lebih baik.
